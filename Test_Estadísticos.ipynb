Las 6 primeras pel√≠culas tienen un promedio mayor que el promedio de todas las pel√≠culas.

```python
#Analizamos como se comporta el acumualado de los promedios
import matplotlib.pyplot as plt

promedios = list()
for i in range(1, len(nota_promedio_por_pelicula_10_o_mas_votos)):
  promedios.append(nota_promedio_por_pelicula_10_o_mas_votos[0:i].mean())

promedios

plt.plot(promedios)
```

Los primeros registros muestran una oscilaci√≥n muy grande, en la medida en que crecen los datos la serie se comienza a estabilizar, cuanto m√°s gande es la muestra, los datos son m√°s consistentes.

```python
#Validamos si el comportamiento es el mismo otorgando un orden aleatorio
import numpy as np

np.random.seed(75243)#Colocamos una semilla, un parametro fijo de 75243
temp =nota_promedio_por_pelicula_10_o_mas_votos.sample(frac=1)

promedios = list()
for i in range(1, len(temp)):
  promedios.append(temp[0:i].mean())

promedios

plt.plot(promedios)
```

Al ordenar el conjunto de datos utilizando el c√≥digo anterior, se observa que los valores se estabilizan a partir de 500, en lugar de a partir de 1000 como en el gr√°fico previo. Esto sugiere que, adem√°s del n√∫mero de muestras, el orden de los datos tambi√©n puede influir en los resultados. Si se extraen muestras de datos no aleatorios, por ejemplo, ordenados de mayor a menor, la probabilidad de tomar decisiones err√≥neas aumenta significativamente. Por ello, es fundamental utilizar datos aleatorios para garantizar un an√°lisis adecuado de las tendencias y la consistencia de la informaci√≥n.


### 3.2 Intervalo de confianza

```python
nota_promedio_por_pelicula_10_o_mas_votos.mean()
```

```python
#Analizando Test Estad√≠sticos Z Test
from statsmodels.stats.weightstats import zconfint # PAra calcular el intervalo de confianza con el Test Z usamos 'zconfint'

zconfint(nota_promedio_por_pelicula_10_o_mas_votos)
```

El resultado muestra que hay un 5% de probabilidad de que, al seleccionar un dato aleatorio del universo, no se encuentre dentro del intervalo de confianza (3.422668132218516, 3.464953011274652). Por lo tanto, existe un 95% de confianza en que, al elegir un dato aleatorio, este estar√° dentro de dicho intervalo.
Se recomienda usar Z Tets con volumenes grandes de datos y T Test con volumenes m√°s cortos de datos

```python
#Analizando Test Estad√≠sticos  T Test
from statsmodels.stats.weightstats import DescrStatsW # PAra calcular el intervalo de confianza con el Test T usamos 'DescrStatsW'

descr_notas_10_votos= DescrStatsW(nota_promedio_por_pelicula_10_o_mas_votos)
descr_notas_10_votos.tconfint_mean()
```

Al usar 'tconfint_mean' nos devuelve los intervalos de confianza, cuyos valores son parecidos al anterior m√©todo

```python
#Calculando  Z Test y  T Test usando 'DescrStatsW'
from statsmodels.stats.weightstats import DescrStatsW # PAra calcular el intervalo de confianza con el Test T usamos 'DescrStatsW'

descr_notas_10_votos= DescrStatsW(nota_promedio_por_pelicula_10_o_mas_votos)
descr_notas_10_votos.tconfint_mean()
```

```python
#Calculando  Z Test y  T Test usando 'DescrStatsW'
descr_notas_10_votos.zconfint_mean()
```
