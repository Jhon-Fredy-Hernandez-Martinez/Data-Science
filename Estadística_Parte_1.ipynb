### Paso 1 - Especifique los límites de cada clase

Utilice la siguiente clasificación:

<b>A</b> ► Más de 20 SM

<b>B</b> ► De 10 a 20 SM

<b>C</b> ► De 4 a 10 SM

<b>D</b> ► De 2 a 4 SM

<b>E</b> ► Hasta 2 SM

donde <b>SM</b> es el valor del salario mínimo en ese momento. En nuestro caso <b>788 mil pesos colombianos</b> (2018):

<b>A</b> ► Más de 15.760

<b>B</b> ► De 7.880 a 15.760

<b>C</b> ► De 3.152 a 7.880

<b>D</b> ► De 1.576 a 3.152

<b>E</b> ► Hasta 1.576


```python
# Determinar el Ingreso mínimo de la muestra
datos.Ingreso.min()
```

```python
# Determinar el Ingreso máximo de la muestra
datos.Ingreso.max()
```

```python
# Los Limites van entre 0 y 200.000
# Creamos los intervalos de la clase
clases =[0,1576, 3152, 7880, 15760, 200000]
```

```python
# Creamos el nombre que va a tomar cada uno de los datos es decir las etiquetas
labels = ['E','D', 'C', 'B', 'A']
```

### Paso 2 - Crear la tabla de frecuencias

Utilizamos la función cut()

La función cut() en Python es una herramienta útil en la biblioteca pandas y se usa principalmente para dividir un conjunto de datos continuos en intervalos discretos. Esto es útil cuando deseas agrupar datos en categorías o "bins" (contenedores) en lugar de manejar cada valor individualmente.

https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.cut.html

```python
# Hacer una categorización. Es decir se coloca un intervalo a la clase
frecuencia = pd.value_counts (pd.cut(x = datos.Ingreso,
       bins = clases, # Establecer los limites
       labels = labels, # Establecer las etiquetas
       include_lowest = True) # Incluye el limite inferior de cada categoria
)
frecuencia
```

```python
#Observamos que las 3 primeras observaciones estan clsificadas como ingreso E
datos.head()
```

```python
# Analizar los porcentajes con 'normalize = True'

porcentaje = pd.value_counts (pd.cut(x = datos.Ingreso,
       bins = clases, # Establecer los limites
       labels = labels, # Establecer las etiquetas
       include_lowest = True), # Incluye el limite inferior de cada categoria
        normalize= True   # Genera la proporción con respecto al total para calcular porcentajes
)*100
porcentaje
```

```python
dist_frec_cuantitativa_personalizadas = pd.DataFrame({'Frecuencia': frecuencia, 'Porcentaje (%)': porcentaje})
dist_frec_cuantitativa_personalizadas
```

```python
# Ordenar las clases con la función 'sort_index()
dist_frec_cuantitativa_personalizadas.sort_index(ascending = False)
```

### Ejercicio
Cuando construimos distribuciones de frecuencia con variables cuantitativas, necesitamos, en la mayoría de los casos, definir las clases en las que queremos agrupar nuestros datos. La biblioteca pandas nos ofrece la función cut() para realizar esta tarea.

Revise el siguiente código:

```python
clases = [datos.Altura.min(), 1.65, 1.75, datos.Altura.max()]
labels = ['1 - Baja', '2 - Media', '3 - Alta']

frecuencia = pd.value_counts(
    pd.cut(
        x = datos.Altura,
        bins = clases,
        labels = labels,
        include_lowest = True
    )
)

porcentaje = pd.value_counts(
    pd.cut(
        x = datos.Altura,
        bins = clases,
        labels = labels,
        include_lowest = True
    ), normalize = True
) * 100

dist_frec_altura = pd.DataFrame(
    {'Frecuencia': frecuencia, 'Porcentaje (%)': porcentaje}
)

dist_frec_altura.rename_axis('Estatura', axis= 'columns', inplace = True)

dist_frec_altura.sort_index(ascending = True, inplace = True)

dist_frec_altura
```

## <font color=green>2.3 Distribución de frecuencia para variables cuantitativas (clases de amplitud fija)</font>
***

### Importando bibliotecas

http://www.numpy.org/

```python
import numpy as np
```

### Paso 1 - Difiniendo el número de clases

#### Regla de Sturges

# $$k = 1 + \frac {10}{3}\log_{10}n$$

```python
# Obtenemos el tamaño de la muestra con 'shape[0]'
n= datos.shape[0]
n
```

```python
# Utilizamos la Regla de Sturges para obtener el número óptimo de clases
k = 1+(10/3)*np.log10(n)
k
```

```python
# Redondeamos las clases a cero decimales
round(k)
```

### Paso 2 - Crear la tabla de frecuencias

```python
#Obtenemos la tabla de frecuencia, recordemos que value_counts ordena de acuerdo a la frecuencia de cada clase
frecuencia = pd.value_counts(
    pd.cut(
        x = datos.Ingreso,
        bins = round(k),
        include_lowest = True
    )
)
frecuencia
```

```python
#Ordenamos de acuerdo a los intervalos con 'sort = False'
frecuencia = pd.value_counts(
    pd.cut(
        x = datos.Ingreso,
        bins = round(k),
        include_lowest = True
    ),
    sort = False
)
frecuencia
```

```python
#Construimos la tabla de porcentajes
porcentaje = pd.value_counts(
    pd.cut(
        x = datos.Ingreso,
        bins = round(k),
        include_lowest = True
    ),
    sort = False,
    normalize= True
)*100
porcentaje
```

```python
# Colocamos los bnombres a las columnas
dist_frec_cuantitativa_amplitud_fija = pd.DataFrame({'Frecuencia': frecuencia, 'Porcentaje (%)': porcentaje})
dist_frec_cuantitativa_amplitud_fija
```

## <font color=green>3.2 Mediana</font>
***

Para obtener la mediana de un conjunto de datos, debemos proceder de la siguiente manera:
1. Ordenar el conjunto de datos;
2. Identificar el número de observaciones (registros) del conjunto de datos ($n$);
3. Identicar el elemento del medio:

> Cuando $n$ sea impar, la posición del elemento del medio se obtendrá de la siguiente manera:


# $$Elemento_{Md} = \frac{n+1}{2}$$

> Cuando $n$ sea par, la posición del elemento mediano se obtendrá de la siguiente manera:


# $$Elemento_{Md} = \frac{n}2$$

4. Obtener la mediana:

> Cuando $n$ sea impar:


# $$Md = X_{Elemento_{Md}}$$

> Cuando $n$ sea par:


# $$Md = \frac{X_{Elemento_{Md}} + X_{Elemento_{Md}+1}}2$$
***

# <font color=green>4 MEDIDAS DE LOCALIZACIÓN</font>
***

Las medidas de localización en estadística son valores que indican la posición de los datos dentro de una distribución, ayudando a identificar su centro, puntos de dispersión y ciertos percentiles. Estas medidas son fundamentales para resumir y comprender un conjunto de datos y son usadas comúnmente en ciencia de datos para entender la distribución de las variables y tomar decisiones basadas en la ubicación relativa de los datos. A diferenia de las medidas de tendencia central, las medidas de localizacion no se ven influenciadas por la presencia de valores extremos.

## <font color=green>4.1 Cuartiles, deciles y percentiles</font>
***

Hay una serie de medidas de posición similares en su diseño a la mediana, aunque no son medidas de tendencia central. Como se sabe, la mediana divide la distribución en dos partes iguales en términos del número de elementos en cada parte. Los cuartiles, por otro lado, permiten que la distribución se divida en cuatro partes iguales en términos del número de elementos en cada uno; deciles en diez partes y centiles en cien partes iguales.

Cuartiles: Dividen los datos en cuatro partes.
𝑄1 Q1, 𝑄2 Q2 (mediana) y 3

***
La DM calcula el promedio de las diferencias absolutas entre cada valor y la media, es decir, no se tienen en cuenta los signos de las diferencias, solo las magnitudes.

# $$DM = \frac 1n\sum_{i=1}^{n}|X_i-\bar{X}|$$



```python
df.mean()
# María y Pablo tienen el mismo promedio. Sin embargo, Pablo tiene mejores notas todas las materias mientras que el resultado de Maria obedece a una mayor Dispersión y Concentracón de las notas
#ya que no tiene notas buenas en todas las materias
# Por esta razón, las medidas de tendencia central no son suficientes para mostrar la relidad de los datos
# Esto justifica el uso medidas de Dispersión
```

```python
df.median()
# MAria y Pablo tiene la misma mediana.
```

```python
notas_maria = df[['María']] # Con el doble corchete convertimos a DataFrame
notas_maria
```

```python
nota_media_maria = notas_maria.mean()[0] # sacamos solo el número con [0]
nota_media_maria
```

## <font color=green>5.2 Varianza</font>
***

### Varianza

La varianza se construye a partir de las diferencias entre cada observación y la media de los datos, es decir, la desviación alrededor de la media. Al calcular la varianza, las desviaciones alrededor de la media son elevadas al cuadrado.

La varianza es una medida estadística que indica el grado de dispersión o variabilidad de un conjunto de datos en relación con su media. En otras palabras, muestra cuánto se desvían los valores individuales de un conjunto respecto al valor promedio (media). Se calcula como el promedio de los cuadrados de las diferencias entre cada dato y la media.

La **varianza de la población** se calcula usando el número total de elementos \(n\) y se aplica cuando se tiene acceso a todos los datos de la población, sin necesidad de correcciones. En cambio, la **varianza de la muestra** se calcula utilizando \(n-1\) en el denominador, un ajuste conocido como corrección de Bessel, que se utiliza cuando se trabaja con una muestra representativa de la población. Este ajuste compensa el sesgo que surge al estimar la varianza de la población a partir de una muestra, garantizando que la estimación sea más precisa y **insesgada**.

### Varianza de la población

# $$\sigma^2 = \frac 1n\sum_{i=1}^{n}(X_i-\mu)^2$$



### Varianza de la muestra

# $$S^2 = \frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$$

```python
# LAs desviaciones se elevan al cuadrado con la función .pow()
notas_maria['(Desviación^2)']= notas_maria['Desviación'].pow(2)
notas_maria
```

```python
#Calculo Manual
#La Varianza de los datos es:
notas_maria['(Desviación^2)'].sum() / (len(notas_maria)-1)
```

```python
#Calculo por mediode la función
notas_maria['María'].var()
```

## <font color=green>5.3 Desviación estándar</font>
***
La DE, por otro lado, calcula el promedio de las diferencias cuadradas entre cada valor y la media, lo que amplifica el impacto de las observaciones más alejadas de la media (es decir, los valores atípicos tienen un mayor peso).




Una de las restricciones de la varianza es el hecho de que proporciona medidas cuadráticas de las unidades originales; la varianza de las medidas de longitud, por ejemplo, está en unidades de área. Por lo tanto, el hecho de que las unidades sean diferentes dificulta la comparación de la dispersión con las variables que la definen. Una forma de eliminar esta dificultad es considerar su raíz cuadrada.

### Desviación estándar de la población

# $$\sigma = \sqrt{\frac 1n\sum_{i=1}^{n}(X_i-\mu)^2} \Longrightarrow \sigma = \sqrt{\sigma^2}$$


### Desviación estándar de la muestra

# $$S = \sqrt{\frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2} \Longrightarrow S = \sqrt{S^2}$$

```python
varianza = notas_maria['María'].var()
varianza
```

```python
# Forma Manual
#La Desviación Estandar es la Raíz Cuadrada de la Varianza
np.sqrt(varianza)
```

```python
# Forma Automatica
desviacion_estandar = notas_maria['María'].std()
desviacion_estandar
```
